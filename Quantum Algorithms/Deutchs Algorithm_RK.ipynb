```python
import pennylane as qml
from pennylane import numpy as np
```

```python
dev = qml.device("default.qubit", wires=2,shots=100)
```

```python
@qml.qnode(dev)
def QA(oracle):
    qml.PauliX(wires=1)
    qml.Hadamard(wires=0)
    qml.Hadamard(wires=1)
    oracle()
    qml.Hadamard(wires=0)
    return qml.probs(wires=0)

def oracle1():
    qml.CNOT(wires=[1,0])
```

```python
res=QA(oracle1)
print(res)
```

```python
cd=qml.draw_mpl(QA)(oracle=oracle1)
print(cd)
```
